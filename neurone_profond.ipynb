```python
# importation des bibliothéques pour etablissement reseau neurone
from tensorflow.keras.models import Sequential
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
```

```python
# Division de nos donnes en cible et features
features=bank_neurone.drop(['deposit'], axis=1)
labels=bank_neurone['deposit']
```

```python
# encodage des features avec methode astype.cat.codes
for i in features.select_dtypes(include='object').columns:
    features[i]=features[i].astype('category').cat.codes
```

```python
features.head()
```

```python
# transformation de notres cible en tableau numpy
feature_np=features.to_numpy()
labels_np=labels.to_numpy()
feature_np.shape
labels_np.shape
```

```python
# redimentionnemen tde notre feature
labels_np=labels_np.reshape(-1, 1)
```

```python
# normalisation des features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
features_scaler = scaler.fit_transform(feature_np)

features_scaler
```

```python
# encodage one hot de la cible
encoder = OneHotEncoder(sparse_output=False)
encoded_labels = encoder.fit_transform(labels_np)
encoded_labels
```

```python
# Split the dataset into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(features_scaler, 
                encoded_labels, test_size=0.2, random_state=42)
```

```python
# Build the model
model = Sequential([
    # First hidden layer: 2 neurons with 'relu' activation for non-linear patterns
    Dense(units=10, activation='relu', input_dim=16),
    # Additional hidden layers: increasing capacity to learn complex representations
    Dense(units=16, activation='relu'),
    Dense(units=16, activation='relu'),
    Dense(units=16, activation='relu'),
    Dense(units=16, activation='relu'),
    Dense(units=16, activation='relu'),
    Dense(units=16, activation='relu'),
    # Output layer: 1 neuron with 'softmax' activation to output a probability
    Dense(units=2, activation='softmax')])
```

```python
# Compile the model
optimizer = Adam(learning_rate=0.0005)
model.compile(optimizer=optimizer, 
loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
# Train the model
history=model.fit(x_train, y_train, epochs=100, batch_size=10)
```

```python
# Evaluate the model on the test set
loss, accuracy = model.evaluate(x_test, y_test)
print(f'Test Loss: {loss}')
print(f'Test Accuracy: {accuracy}')
```

```python
"""conclusion: overfitting est present car accuracy entrainement est superieur a accuracy test"""
```

```python
# Plot training & validation accuracy and loss values
plt.figure(figsize=(15,10))
plt.plot(history.history['accuracy'], label='Accuracy')  # Shows how accuracy improves
plt.plot(history.history['loss'], label='Loss')          # Shows how loss decreases

plt.title('Model Training Performance')
plt.ylabel('Value')
plt.xlabel('Epoch')
plt.legend(loc="upper left")
plt.show()
```

```python
"""a partir de epoch 20, on remarque que accuracy stagne, indiquant que le modèle a peut-être atteint sa capacité maximale d'apprentissage sur les données d'entraînement. Cela suggère également que le modèle pourrait bénéficier de techniques supplémentaires pour améliorer sa généralisation, telles que la régularisation ou l'augmentation des données."""
```

```python
# importation de la couche Dropout
from tensorflow.keras.layers import Dropout
```

```python
model_drop = Sequential([
    Dense(units=10, activation='relu', input_dim=16),
    Dropout(0.2),
    Dense(units=16, activation='relu'),
    Dropout(0.16),

    Dense(units=16, activation='relu'),
    Dropout(0.12),
    Dense(units=16, activation='relu'),
    Dropout(0.08),
    Dense(units=16, activation='relu'),
    Dropout(0.04),
    Dense(units=16, activation='relu'),
    Dropout(0.02),
    Dense(units=16, activation='relu'),
    # Output layer: 1 neuron with 'softmax' activation to output a probability
    Dense(units=2, activation='softmax')])
```

```python
# Compilation du model avec Dropout
model_drop.compile(optimizer=Adam(learning_rate=0.0005), 
loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
import tensorflow as tf
tf.config.run_functions_eagerly(True)
tf.compat.v1.enable_eager_execution()
```

```python
# Entrainement du model avec Dropout
history_drop=model_drop.fit(x_train, y_train, epochs=17, batch_size=10)
```

```python
loss, accuracy = model_drop.evaluate(x_test, y_test)
print(f'Test Loss: {loss}')
print(f'Test Accuracy: {accuracy}')
```
